```python
from scipy.signal import gaussian

thz2cm=33.305
data = np.genfromtxt('md-ir-spectra-4000.dat')
x=data[:,0]
y=data[:,1]
# Define the broadening factor (standard deviation)
broadening_factor = 2

# Create a Gaussian kernel
window = int(broadening_factor * 10)  # Adjust the window size as needed
gaussian_window = gaussian(window, std=broadening_factor)

# Apply convolution to the FFT spectrum
y_broadened = np.convolve(y, gaussian_window, mode='same')

plt.plot(x*thz2cm,y_broadened)
plt.xlim([400,800])
plt.ylim([0,0.2])
plt.show()
```
